e) Сгенерируйте выборку объема n = 100 для некоторого $\theta$. Вычислить дов. интервалы для $\beta = 0.95$

```python
import numpy as np
from rich.console import Console
from rich.table import Table
from rich.panel import Panel
from rich.text import Text

n = 100
theta = 1000
beta = 0.95

def ro(x : float, theta : float):
    return 1 / theta if (2 * theta >= x >= theta) else 0

def distr(x : float, theta : float):
    return x / theta if (2 * theta >= x >= theta) else 0

def inv_distr(y : float, theta : float):
    return theta * (y + 1)

randoms = np.random.random(n)
sample = np.array([inv_distr(y, theta) for y in randoms])

intervals_width: dict[str, float] = {}

Console = Console()
Table1 = Table()
Table1.add_column(header="Выборка", justify="center")

for i in range(len(sample) // 4):
  s = sample[i]
  s_1 = sample[i + 1]
  s_2 = sample[i + 2]
  s_3 = sample[i + 3]

  Table1.add_row(f"{s:.6f} │ {s_1:.6f} │ {s_2:.6f} │ {s_3:.6f}")

Console.print(Table1)
```

```python
lower = np.max(sample) / (1 + ((1 + beta) / 2) ** (1 / n))
upper = np.max(sample) / (1 + ((1 - beta) / 2) ** (1 / n))

Panel1 = Panel(Text(f"{lower} < θ < {upper}\n\tl = {upper - lower}"), title="Точный доверительный интервал")

Console.print(Panel1, justify="left")

intervals_width["Точный"] = upper - lower
```

```python
S = (np.sum([(x - np.mean(sample)) ** 2 for x in sample])) / (n - 1)

lower = 2 * np.mean(sample) / 3 - 3.92 * np.sqrt(S * (n - 1)) / (3 * n)
upper = 2 * np.mean(sample) / 3 + 3.92 * np.sqrt(S * (n - 1)) / (3 * n)

Panel2 = Panel(Text(f"{lower} < θ < {upper}\n\tl = {upper - lower}"), title="Асимптотический доверительный интервал")

Console.print(Panel2, justify="left")

intervals_width["Асимптотический"] = upper - lower
```

g) Численно постройте доверительный интервал через bootstrap

```python
bootstrap_iter = 1000
omp_mark = (n + 1) * np.max(sample) / (2 * n + 1)

bootstrap_delta: list[float] = []

for _ in range(bootstrap_iter):
  bootstrap_delta.append((n + 1) * np.max(np.random.choice(sample, size=n)) /
                         (2 * n + 1) - omp_mark)
  
lower = omp_mark - sorted(bootstrap_delta)[int((1 + beta) * bootstrap_iter / 2)]
upper = omp_mark - sorted(bootstrap_delta)[int((1 - beta) * bootstrap_iter / 2)]

Panel3 = Panel(Text(f"{lower} < θ < {upper}\n\tl = {upper - lower}"), title="Доверительный интервал через bootstrap")

Console.print(Panel3, justify="left")

intervals_width["Bootstrap"] = upper - lower
```

e) Сравнить доверительные интервалы

```python
Table2 = Table()
Table2.add_column("Тип интервала")
Table2.add_column("Ширина в порядке увеличения")
for pair in sorted(intervals_width.items(), key= lambda x : x[1]):
    Table2.add_row(str(pair[0]), str(pair[1]))

Console.print(Table2)
```
