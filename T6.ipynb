d) Сгенерируйте выборку объема n = 100 для некоторого $\theta$. Вычислить дов. интервалы для $\beta = 0.95$

```python
import numpy as np
from rich.console import Console
from rich.table import Table
from rich.panel import Panel
from rich.text import Text

n = 100
theta = 1000
beta = 0.95

def ro(x: float, theta: float):
  return (theta - 1) / (x ** theta) if x >= 1 else 0


def distib(x: float, theta: float):
  return (1 - x ** (1 - theta)) if x >= 1 else 0


def inv_distr(y: float, theta: float):
  return (1 - y) ** (1 / (1 - theta))

randoms = np.random.random(n)
sample = np.array([inv_distr(y, theta) for y in randoms])
ln_mean = np.mean([np.log(x) for x in sample])

intervals_width: dict[str, float] = {}

Console = Console()
Table1 = Table()
Table1.add_column(header="Выборка", justify="center")

for i in range(len(sample) // 4):
  s = sample[i]
  s_1 = sample[i + 1]
  s_2 = sample[i + 2]
  s_3 = sample[i + 3]

  Table1.add_row(f"{s:.6f} │ {s_1:.6f} │ {s_2:.6f} │ {s_3:.6f}")

Console.print(Table1)
```

```python
lower = 2 ** ln_mean - (1.96 * np.log(2) * 2 ** ln_mean * ln_mean) / np.sqrt(n)
upper = 2 ** ln_mean + (1.96 * np.log(2) * 2 ** ln_mean * ln_mean) / np.sqrt(n)

Panel1 = Panel(Text(f"{lower} < med < {upper}\n\tl = {upper - lower}"), title="Асимптотический доверительный интервал для медианы")

Console.print(Panel1, justify="left")
```

```python
lower = 1 - (1.96 - np.sqrt(n)) / (ln_mean * np.sqrt(n))
upper = 1 + (1.96 + np.sqrt(n)) / (ln_mean * np.sqrt(n))

Panel2 = Panel(Text(f"{lower} < θ < {upper}\n\tl = {upper - lower}"), title="Асимптотический доверительный интервал для θ")

Console.print(Panel2, justify="left")

intervals_width["Асимптотический для θ"] = upper - lower
```

e) Численно постройте доверительный интервал через bootstrap параметрически и непараметрически

```python
bootstrap_iter = 1000
omp_mark = 1 + 1 / ln_mean

bootstrap_delta: list[float] = []

for _ in range(bootstrap_iter):
    bootstrap_delta.append(1 + 1 / np.mean([np.log(x) for x in np.random.choice(sample, size=n)]).astype(float) - omp_mark)

lower = omp_mark - sorted(bootstrap_delta)[int((1 + beta) * bootstrap_iter / 2)]
upper = omp_mark - sorted(bootstrap_delta)[int((1 - beta) * bootstrap_iter / 2)]

Panel3 = Panel(Text(f"{lower} < θ < {upper}" f"\n\tl = {upper - lower}", style="bold"), title="Непараметрический bootstrap'овский доверительный интервал")

intervals_width["Непараметрический bootstrap"] = upper - lower

Console.print(Panel3, justify="left")
```

```python
bootstrap_iter = 50000
bootstrap_delta: list[float] = []

for _ in range(bootstrap_iter):
    random_sample = []

    for _ in range(n):
        y = np.random.random()
        random_sample.append(inv_distr(y, omp_mark))

    bootstrap_delta.append(1 + 1 / np.mean([np.log(x) for x in random_sample]) - omp_mark)

lower = omp_mark - sorted(bootstrap_delta)[int((1 + beta) * bootstrap_iter / 2)]
upper = omp_mark - sorted(bootstrap_delta)[int((1 - beta) * bootstrap_iter / 2)]

Panel4 = Panel(Text(f"{lower} < θ < {upper}" f"\n\tl = {upper - lower}", style="bold"), title="Параметрический bootstrap'овский доверительный интервал")

intervals_width["Параметрический bootstrap"] = upper - lower

Console.print(Panel4, justify="left")
```

e) Сравнить доверительные интервалы

```python
Table2 = Table()
Table2.add_column("Тип интервала")
Table2.add_column("Ширина в порядке увеличения")
for pair in sorted(intervals_width.items(), key= lambda x : x[1]):
    Table2.add_row(str(pair[0]), str(pair[1]))

Console.print(Table2)
```
